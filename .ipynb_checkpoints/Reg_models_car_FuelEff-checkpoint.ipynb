# Creating regression models for car fuel efficiency.

- The goal of this homework is to create a regression model for predicting the car fuel efficiency (column 'fuel_efficiency_mpg').

```python
import pandas as pd
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

from sklearn.ensemble import RandomForestRegressor
```

### **Data cleaning and preparation**

- Downloading the dataset
- filling up missing values with zeros

```python
df = pd.read_csv("car_fuel_efficiency.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
df.head()
```

## **Setting up the train/val/test framework**

- We first do train/validation/test split with 60%/20%/20% distribution
- Then, apply one-hot encoding to categorical features and get the feature matrix

```python
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=1)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.fuel_efficiency_mpg.values
y_val = df_val.fuel_efficiency_mpg.values
y_test = df_test.fuel_efficiency_mpg.values
```

```python
del df_train['fuel_efficiency_mpg']
del df_val['fuel_efficiency_mpg']
del df_test['fuel_efficiency_mpg']
```

```python
len(df_train),len(df_val), len(df_test)
```

### Using DictVectorizer to setup one hot encoding

```python
train_dicts = df_train.to_dict(orient='records')
```

```python
dv = DictVectorizer(sparse=True)

##sparse=True (default): Returns a sparse matrix, which is memory-efficient for data with many zeros. It only stores non-zero values and their positions.
##sparse=False: Returns a regular dense NumPy array, storing all values including zeros.

X_train = dv.fit_transform(train_dicts)
```

```python
dv.get_feature_names_out()
```

```python
feature_names = dv.get_feature_names_out()
```

## **1. Let's train a decision tree regressor to predict the fuel_efficiency_mpg variable.**


- Train a model with max_depth=1.
- Which feature is used for splitting the data?

```python
dt = DecisionTreeRegressor(max_depth=1)
dt.fit(X_train, y_train)
```

```python
# Check which feature is used at the root node (first split)
feature_index = dt.tree_.feature[0]
feature_name = feature_names[feature_index]

print(f"Feature used for splitting: {feature_name}")
```

**Alternatively to check for the node for spliting**

```python
# Get feature importances
importances = dt.feature_importances_
feature_names = dv.get_feature_names_out()

# Find the feature with non-zero importance (the one used for splitting)
for name, importance in zip(feature_names, importances):
    if importance > 0:
        print(f"Split feature: {name} (importance: {importance})")
```

**completing simple decision tree regression(dt) training by validating on val dataset to the RMSE**

```python
val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
y_val_pred = dt.predict(X_val)

##NB: Why the difference?

#Classifiers (predict_proba): Return probabilities for each class (e.g., 70% chance of class 1, 30% chance of class 0)
#Regressors (predict): Return continuous numerical predictions (e.g., 42.5, 103.7, etc.)
```

```python
# Root Mean Squared Error (most common)
rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
print(f"RMSE: {rmse}")
```

## **2. Train a random forest regressor with these parameters:**

- n_estimators=10
- random_state=1
- n_jobs=-1 (optional - to make training faster)

What's the RMSE of this model on the validation data?

```python
rf = RandomForestRegressor(n_estimators=10, random_state=1,n_jobs=-1)
rf.fit(X_train, y_train)
```

```python
y_val_pred = rf.predict(X_val)
```

```python
rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
print(f"RMSE: {rmse}")
```

## **3. Now let's experiment with the n_estimators parameter**

- Try different values of this parameter from 10 to 200 with step 10.
- Set random_state to 1.
- Evaluate the model on the validation dataset.

After which value of n_estimators does RMSE stop improving? Consider 3 decimal places for calculating the answer.

```python
scores =[]

for n in range (10,201,10):
    rf = RandomForestRegressor(n_estimators=n, random_state=1,n_jobs=-1)
    rf.fit(X_train, y_train)

    y_val_pred = rf.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_val_pred)).round(3)
    scores.append((n, rmse))
```

```python
scores
```

```python
columns = ['n_estimators', 'rmse']
df_scores =pd.DataFrame(scores, columns = columns)
```

```python
plt.plot(df_scores.n_estimators, df_scores.rmse)

plt.xlabel('n_estimator')
plt.ylabel('RMSE')
```

## **4. Let's select the best max_depth:**

- Try different values of max_depth: [10, 15, 20, 25]
- For each of these values,
- try different values of n_estimators from 10 till 200 (with step 10)
- calculate the mean RMSE
- Fix the random seed: random_state=1
  
What's the best max_depth, using the mean RMSE?

```python
scores =[]

for d in [10, 15, 20, 25]:
    
    for n in range (10,201,10):
        rf = RandomForestRegressor(n_estimators=n,
                                   max_depth = d,
                                   random_state=1, 
                                   n_jobs=-1)
        rf.fit(X_train, y_train)
    
        y_val_pred = rf.predict(X_val)
        rmse = np.sqrt(mean_squared_error(y_val, y_val_pred)).round(3)
        scores.append((d, n, rmse))
```

```python
columns = ['max_depth', 'n_estimators', 'rmse']
df_scores =pd.DataFrame(scores, columns = columns)

# To see the top results (lowest RMSE) first:
df_scores_sorted = df_scores.sort_values('rmse', ascending=True)
df_scores_sorted.head()
```

```python
for d in [10, 15, 20, 25]:
    df_subset = df_scores[df_scores.max_depth == d]
    plt.plot(df_subset.n_estimators, df_subset.rmse, label = 'max_depth =%d' % d)


plt.legend()
plt.xlabel('n_estimators') #no of decision trees, or subset of data used by dt for training.
plt.ylabel('rmse')
```

## **5 We can extract feature importance information from tree-based models.**

At each step of the decision tree learning algorithm, it finds the best split. When doing it, we can calculate "gain" - the reduction in impurity before and after the split. This gain is quite useful in understanding what are the important features for tree-based models.

In Scikit-Learn, tree-based models contain this information in the feature_importances_ field.

For this homework question, we'll find the most important feature:

Train the model with these parameters:
- n_estimators=10,
- max_depth=20,
- random_state=1,
- n_jobs=-1 (optional)

Get the feature importance information from this model
What's the most important feature?

```python
rf = RandomForestRegressor(n_estimators=10, 
                           max_depth = 20,
                           random_state=1,
                           n_jobs=-1)
rf.fit(X_train, y_train)
    
# Get feature importances
importances = rf.feature_importances_
feature_names = dv.get_feature_names_out()

# Find the feature with non-zero importance (the one used for splitting)
for name, importance in zip(feature_names, importances):
    if importance > 0:
        print(f"Split feature: {name} (importance: {importance})")

#y_val_pred = rf.predict(X_val)
#rmse = np.sqrt(mean_squared_error(y_val, y_val_pred)).round(3)
```

## **6. Now let's train an XGBoost model! For this question, we'll tune the eta parameter:**

- Install XGBoost
- Create DMatrix for train and validation
- Create a watchlist
- Train a model with these parameters for 100 rounds:

```python
import xgboost as xgb
```

```python
features = dv.get_feature_names_out().tolist()
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
watchlist = [(dtrain, 'train'), (dval, 'val')]
```

```python
%%capture output 
#this will contain the output or result. So we don't have all the values printed out on the screen.

xgb_params = {
    'eta': 0.1, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}

model = xgb.train(xgb_params, dtrain, num_boost_round = 100,
                 verbose_eval = 10, #Print progress every 10 iterations
                 evals=watchlist)
```

```python
y_val_pred = model.predict(dval)
```

```python
rmse = np.sqrt(mean_squared_error(y_val, y_val_pred)).round(3)
rmse
```
